`np.linspace(-5,5,100)`: Return evenly spaced numbers over a specified interval.
`plt.plot(x,y)`: Plot a simple x,y graph

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,100)
y = tf.keras.activations.sigmoid(x)
# y = [1/(1+np.exp(-xi)) for xi in x] # Using in-built package for sigmoid function

# Plot the graph
plt.plot(x,y)
plt.title('Activation: Sigmoid')
plt.xlabel('x')
plt.ylabel('Sigmoid (x)')
plt.grid() # show a grid
plt.show()
```

```python
x = np.linspace(-5,5,100)
y = tensorflow.keras.activations.relu(x)

plt.plot(x,y)
plt.title('Activation: ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid()
plt.show()
```
